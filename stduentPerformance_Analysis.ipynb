Basic analysis of student marks using Python (Pandas & Matplotlib).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("studentMarks.csv")
```

```python
df.head()
```

```python
df["Total"]= df["Maths"]+df["Science"]+df["English"]
df["Average"]=round(df["Total"]/3,2)
df.head()
```

```python
topper=df.loc[df["Total"].idxmax()]
topper
```

```python
sub_avg=df[["Maths","Science","English"]].mean()
sub_avg
```

```python
print("overall class average:" ,df["Average"].mean())
```

```python
sub_avg.idxmax()
```

```python
def grade(avg):
    if avg>=90:
        return "A"
    elif avg>=80:
        return "B"
    elif avg>=70:
        return "C"
    elif avg>=60:
        return "D"
    else:
        return "F"
df["Grade"]=df["Average"].apply(grade)
df.head()
```

```python
df_sorted = df.sort_values(by="Total", ascending=False)
top5 =df_sorted.head()
top5
```

```python
bottom5=df_sorted.tail()
bottom5
```

```python
df["Grade"].value_counts()
```

```python
sub_avg.plot(kind="bar")
plt.title("Average Marks per Subject")
plt.ylabel("Average Marks")
plt.show()
```

```python
plt.bar(top5["Name"], top5["Total"])
plt.title("Top 5 Students by Total Marks")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.show()
```
